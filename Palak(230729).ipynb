```python
stocks = ["TCS.NS", "HDFCBANK.NS", "RELIANCE.NS", "ICICIBANK.NS","INFY.NS"]


data = yf.download(stocks, start="2021-01-01", end="2025-06-27", auto_adjust=True)
```

```python
data
```

```python
close = data["Close"]
close
```

```python
close = close[stocks]
close
```

```python
returns = close.pct_change(fill_method=None).dropna(how="all")
returns
```

```python
expectation= returns.mean(axis=0)
expectation
```

```python
covariance = returns.cov()*252
covariance
```

```python


plt.figure(figsize=(10, 8))
sns.heatmap(covariance, annot=True, cmap='coolwarm', fmt=".4f")
plt.title('Covariance Matrix of Stock Returns')
plt.show()
```

```python

try:
    inverse_covariance = np.linalg.inv(covariance)
    print("\nInverse Covariance Matrix:")
    print(inverse_covariance)

    plt.figure(figsize=(10, 8))
    sns.heatmap(inverse_covariance, annot=True, cmap='coolwarm', fmt=".4f")
    plt.title('Inverse Covariance Matrix of Stock Returns')
    plt.show()

except np.linalg.LinAlgError:
    print("Error: The covariance matrix is singular and cannot be inverted.")
```

```python
matrix1 = np.ones(5)
matrix1
```

```python
matrix2 = np.dot(matrix1, np.dot(inverse_covariance, matrix1.T))
matrix2
```

```python
optimal_weight= np.dot(inverse_covariance, matrix1)/matrix2
optimal_weight
```

```python

portfolio_volatility = np.sqrt(np.dot(optimal_weight.T, np.dot(covariance, optimal_weight)))
print(f"Portfolio Volatility: {portfolio_volatility*100:.4f}")
```

Plotting the Markowitz efficient Frontier

```python


def generate_random_weights(num_stocks, num_portfolios):

  all_weights = []
  for _ in range(num_portfolios):
    weights = np.random.random(num_stocks)
    weights /= np.sum(weights)
    all_weights.append(weights)
  return np.array(all_weights)

num_stocks = len(stocks)
num_portfolios = 10000
random_weights = generate_random_weights(num_stocks, num_portfolios)

# You can print or inspect the first few generated weight vectors
print("First 5 random weight vectors:")
print(random_weights[:5])
```

```python

variance_matrix= np.dot(random_weights, np.dot(covariance, random_weights.T))
variance_matrix
```

```python


portfolio_returns = np.dot(random_weights, expectation)
portfolio_volatility = np.sqrt(np.diag(variance_matrix))

plt.figure(figsize=(12, 8))
plt.scatter(portfolio_volatility, portfolio_returns, c=portfolio_returns, cmap='viridis', marker='o')
plt.title('Markowitz Efficient Frontier with Random Portfolios')
plt.xlabel('Portfolio Volatility (Standard Deviation)')
plt.ylabel('Portfolio Expected Return')
plt.colorbar(label='Expected Return')
plt.grid(True)
plt.show()
```
